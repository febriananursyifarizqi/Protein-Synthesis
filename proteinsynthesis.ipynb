```python
def read_seq(inputfile):
    f = open(inputfile, "r") 
    seq = f.read()

    seq = seq.replace("\n", "")  
    seq = seq.replace("\r", "")

    return seq
```

### Transcription

```python
def transcript_dna(seq):
  table = {'A':'U', 'T':'A', 'C':'G', 'G':'C'}
  mrna = ''
  for nucleotide in seq:
    mrna += table[nucleotide]
  return mrna
```

```python
def transcript_rna(seq):  
  mrna = seq.replace("T", "U")
  return mrna
```

### Translation

```python
def translate(mrna):

    table = {
        'AUA':'I', 'AUC':'I', 'AUU':'I', 'AUG':'M',
        'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACU':'T',
        'AAC':'N', 'AAU':'N', 'AAA':'K', 'AAG':'K',
        'AGC':'S', 'AGU':'S', 'AGA':'R', 'AGG':'R',
        'CUA':'L', 'CUC':'L', 'CUG':'L', 'CUU':'L',
        'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCU':'P',
        'CAC':'H', 'CAU':'H', 'CAA':'Q', 'CAG':'Q',
        'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGU':'R',
        'GUA':'V', 'GUC':'V', 'GUG':'V', 'GUU':'V',
        'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCU':'A',
        'GAC':'D', 'GAU':'D', 'GAA':'E', 'GAG':'E',
        'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGU':'G',
        'UCA':'S', 'UCC':'S', 'UCG':'S', 'UCU':'S',
        'UUC':'F', 'UUU':'F', 'UUA':'L', 'UUG':'L',
        'UAC':'Y', 'UAU':'Y', 'UAA':' ', 'UAG':' ',
        'UGC':'C', 'UGU':'C', 'UGA':' ', 'UGG':'W',
    }

    protein = ''  
    for i in range(0, len(mrna), 3): 
        codon = mrna[i:i + 3]
        if len(codon) == 3: 
            protein += table[codon] 
            
    return protein
```

### Contoh Input Output 1: DNA Sequence dari Slide PPT
Proses sintesis protein pada DNA sequence "TCGGTGAATCTGTTTGAT"

```python
gene1 = "TCGGTGAATCTGTTTGAT"
print("gene:", gene1)

mrna1 = transcript_dna(gene1)
print("mrna:", mrna1)

protein1 = translate(mrna1)
print("protein:", protein1)
```

### Contoh Input Output 2: Sars Cov 2 RNA Sequence
Sumber: https://www.ncbi.nlm.nih.gov/nuccore/1798172431 

Materi genetik pada Sars Cov 2 adalah RNA, bukan DNA. Oleh karena itu, proses sintesis protein pada Sars Cov 2 sedikit berbeda. Langkah pertama yang harus dilakukan adalah mengubah setiap basa nitrogen timin (T) menjadi urasil (U). Setelah itu, setiap trikodon akan ditranslasikan menjadi sebuah asam amino.

#### Whole Genome

```python
gene2 = read_seq("sarscov2_genome.txt")
print("gene:", gene2)

mrna2 = transcript_rna(gene2)
print("mrna:", mrna2)

protein2 = translate(mrna2)
print("protein:", protein2)
```

```python
text_file = open(r"whole_protein.txt", "w")
text_file.write(protein2)
text_file.close()
```
